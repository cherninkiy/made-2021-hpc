```python
df = pd.read_csv("result.csv",delimiter=";", names=["implementation", "time", "size", "power", "threads"])
df = df.reset_index(drop=True)
df
```

```python
df_avg = df[["time", "size", "power", "implementation"]] \
    .groupby(by=["size","power", "implementation"]) \
    .mean() \
    .reset_index()

for c in sorted(set(df_avg["size"].values)):
    df1 = df_avg[df_avg["size"] == c]
    display(df1.sort_values(by="time"))
```

```python
df1 = df_avg[df_avg["size"] >= 1000]
display(df1.sort_values(by="time"))
```
