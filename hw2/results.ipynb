```python
df_host = pd.read_csv("host.csv",delimiter=";", names=["runtime", "implementation", "time", "size", "comment"])
df_host["runtime"] = "host"
df_host["comment"] = df_host["runtime"] + " " + df_host["comment"]
df_host
```

```python
df_docker = pd.read_csv("docker.csv",delimiter=";", names=["runtime", "implementation", "time", "size", "comment"])
df_docker["runtime"] = "docker"
df_docker["comment"] = df_docker["runtime"] + " " + df_docker["comment"]
df_docker
```

```python
df = pd.concat((df_host, df_docker), axis=0)
df
```

```python
df_avg = df[["time", "size", "comment", "implementation"]] \
    .groupby(by=["size","comment", "implementation"]) \
    .mean() \
    .reset_index()

for c in sorted(set(df_avg["size"].values)):
    df1 = df_avg[df_avg["size"] == c]
    display(df1.sort_values(by="time"))
```

```python
df1 = df_avg[df_avg["size"] >= 2000]
display(df1.sort_values(by="time"))
```
